## K-NN

```python
# KNN Parameters
k = 3
X = 22  # Input value
```

```python
# Calculate distances (absolute difference in age)
df['distance'] = abs(df['age'] - X)
```

```python
df_sorted = df.sort_values('distance')
```

```python
# Select K nearest neighbors
nearest_neighbors = df_sorted.iloc[:k]
```

```python
# Predict income by taking the average income of K nearest neighbors
predicted_income_knn = nearest_neighbors['income'].mean()
```

```python
df2 = pd.read_csv('weight-height.csv')
df2
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

### Preprocessing 

### Separate Features (X = 'Gender', 'Height') and Target (y = 'Weight')

```python
# Encode 'Gender' into numerical values (0 for Female, 1 for Male)
le = LabelEncoder()
df2['Gender'] = le.fit_transform(df2['Gender'])  # Male = 1, Female = 0

# X = 'Gender' and 'Height', y = 'Weight'
X = df2[['Gender', 'Height']]
y = df2['Weight']
```

```python
df2
```

### Split the Data into Training (70%) and Testing (30%)

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

## Apply Linear Regression

```python
# Create and train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Make predictions
y_pred_train_lr = lr_model.predict(X_train)
y_pred_test_lr = lr_model.predict(X_test)
```

### Evaluate the Linear Regression Model (Training and Testing Accuracy, MSE)

```python
# Training accuracy (R2 score)
train_r2_lr = r2_score(y_train, y_pred_train_lr)
test_r2_lr = r2_score(y_test, y_pred_test_lr)

# Mean Squared Error (MSE) for testing set
mse_test_lr = mean_squared_error(y_test, y_pred_test_lr)
```

```python
print(f"Linear Regression Training Accuracy (R2): {train_r2_lr}")
print(f"Linear Regression Testing Accuracy (R2): {test_r2_lr}")
print(f"Linear Regression MSE on Test Set: {mse_test_lr}")
```

## Apply KNN Regression (with K=3 as an example)

```python
# Create and train the KNN model (with k=3)
knn_model = KNeighborsRegressor(n_neighbors=3)
knn_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_train_knn = knn_model.predict(X_train)
y_pred_test_knn = knn_model.predict(X_test)
```

```python
# Training accuracy (R2 score)
train_r2_knn = r2_score(y_train, y_pred_train_knn)
test_r2_knn = r2_score(y_test, y_pred_test_knn)
```

```python
# Mean Squared Error (MSE) for testing set
mse_test_knn = mean_squared_error(y_test, y_pred_test_knn)

print(f"KNN Training Accuracy (R2): {train_r2_knn}")
print(f"KNN Testing Accuracy (R2): {test_r2_knn}")
print(f"KNN MSE on Test Set: {mse_test_knn}")
```

# Compare Linear Regression And KNN Regression 

```python
print("\n--- Model Comparison ---")
print(f"Linear Regression Testing Accuracy (R2): {test_r2_lr}")
print(f"Linear Regression MSE on Test Set: {mse_test_lr}")

print(f"KNN Testing Accuracy (R2): {test_r2_knn}")
print(f"KNN MSE on Test Set: {mse_test_knn}")
```
